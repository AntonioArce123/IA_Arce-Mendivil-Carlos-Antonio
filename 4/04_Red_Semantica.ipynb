```python
# Tarea 4 – Red Semántica
# Representación sencilla de conceptos y relaciones usando un grafo dirigido

%pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir nodos (conceptos)
conceptos = ["Mamífero", "Animal", "Perro", "Gato", "Ser Vivo", "Doméstico", "Salvaje"]
G.add_nodes_from(conceptos)

# Añadir relaciones (aristas)
relaciones = [
    ("Perro", "Mamífero", "es_un"),
    ("Gato", "Mamífero", "es_un"),
    ("Mamífero", "Animal", "es_un"),
    ("Animal", "Ser Vivo", "es_un"),
    ("Perro", "Doméstico", "tiene_característica"),
    ("Gato", "Doméstico", "tiene_característica"),
    ("Gato", "Salvaje", "puede_ser")
]

for origen, destino, tipo in relaciones:
    G.add_edge(origen, destino, label=tipo)

# Dibujar la red semántica
pos = nx.spring_layout(G)
edge_labels = nx.get_edge_attributes(G, 'label')

plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Red Semántica: Mamíferos y sus relaciones")
plt.show()
```
