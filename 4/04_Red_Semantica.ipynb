```python
# Tarea 4  Red Semntica
# Representacin sencilla de conceptos y relaciones usando un grafo dirigido

%pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Aadir nodos (conceptos)
conceptos = ["Mamfero", "Animal", "Perro", "Gato", "Ser Vivo", "Domstico", "Salvaje"]
G.add_nodes_from(conceptos)

# Aadir relaciones (aristas)
relaciones = [
    ("Perro", "Mamfero", "es_un"),
    ("Gato", "Mamfero", "es_un"),
    ("Mamfero", "Animal", "es_un"),
    ("Animal", "Ser Vivo", "es_un"),
    ("Perro", "Domstico", "tiene_caracterstica"),
    ("Gato", "Domstico", "tiene_caracterstica"),
    ("Gato", "Salvaje", "puede_ser")
]

for origen, destino, tipo in relaciones:
    G.add_edge(origen, destino, label=tipo)

# Dibujar la red semntica
pos = nx.spring_layout(G)
edge_labels = nx.get_edge_attributes(G, 'label')

plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Red Semntica: Mamferos y sus relaciones")
plt.show()
```
